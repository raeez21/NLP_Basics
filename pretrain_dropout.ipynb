```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

Using dropout (using Dropout Class)

```python
prob = .2

dropout = nn.Dropout(p=prob)
x = torch.ones(100)
print('initial data vector:\n ',x,'\n')


y=dropout(x)
print('"Raw" post dropout vector:\n ',y,'\n')
# the non zero activations will be scaled up (that is why we get 1.25)


#normalized
print('normed post dropout vector:\n ',y*(1-prob),'\n')

#how many were dropped
print(f'{100*torch.sum(y==0)/len(x)}% dropped (expected {prob*100:.1f}%)')

# dropout is probablistic, so each time you run this cell different sets of units are dropped
```

```python
x.sum(),y.sum(), (y*(1-prob)).sum()
# as Pytorch internally scaled up non-dropout units, we get sum closer, or else it would be diff (81)
```

How to switch off dropout

```python
#dropout turned off when evaluating model
dropout.eval() #in actual code this will be model.eval()
y = dropout(x)
print(y)
```

DROPOUT USING FUNCTION

```python
F.dropout(x,p=prob)
```

```python
# why 3 methods???
# bcoz of ease of toggling dropout

#dropout used only for training, you want to turn off dropout during evaluation, prediction, production etc
```

Toggling dropout

```python
#dropout turned off when evaluating model
dropout.eval() #in actual code this will be model.eval() [This actually turns off gradient calculation, reg techniques like dropout etc]
y = dropout(x)
print(y)
```

```python
# F.dropout() is not deactivated in eval mode
dropout.eval()
y = F.dropout(x,p=.2)
print(y)
```

```python
# but can you manually swithc it off???

y = F.dropout(x,training=False)
print(y)
```

```python
# model needs to be reset after toggling into eval mode
dropout.train()
y = dropout(x)
print('train() mode on:\n',y,'\n') #with dropout

dropout.eval()
y = dropout(x)
print('eval() mode on:\n',y,'\n')


# dropout.train()
y = dropout(x)
print('eval() remains on:\n',y,'\n') #still w/o dropout
```

```python
# explicitly switch it back to train
dropout.train()
y = dropout(x)
print('eval() remains on:\n',y,'\n')
```

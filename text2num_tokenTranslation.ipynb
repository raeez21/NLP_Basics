```python
import tiktoken
gpt4Tok = tiktoken.get_encoding('cl100k_base')

#bert
from transformers import BertTokenizer
bertTok = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
## attempting a direct translation from GPT4 to BERT and back

startText = 'Hello, my name is Raeez and I like purple.'

#GPT4 token
gpt4toks = gpt4Tok.encode(startText)

berttoks = bertTok.encode(startText)

print(f'start text: \n{startText}')
print(f'GPT 4 tokens: \n{gpt4toks}')
print(f'decoded using GPT4: \n{gpt4Tok.decode(gpt4toks)}')
print(f'decoded using BERT: \n{bertTok.decode(gpt4toks)}')

print(f'\n\n BERT  tokens: \n{berttoks}')
print(f'decoded using BERT \n{bertTok.decode(berttoks)}')
print(f'decoded using GPT4: \n{gpt4Tok.decode(berttoks)}')
```

## Right way to translate

```python
# text -> GPT4 tokens -> text -> BERT tokens

# 1) to GPT tokens
startText = 'Hello, my name is Raeez and I like purple.'
gpt4Toks = gpt4Tok.encode(startText)

# 2) back to text
gpt4ReconText = gpt4Tok.decode(gpt4toks)

# then to bert tokens
bertToks = bertTok.encode(gpt4ReconText)

#4) show the reconstruciton
bertTok.decode(bertToks)

# Note that uncased used in BERT--> so not exactly correct transaltion
```

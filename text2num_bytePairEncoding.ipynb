```python
import numpy as np
```

```python
# some text with lots of repetitions
text = 'like liker love lovely hug hugs hugging hearts'

chars = list(set(text))

chars.sort()
for l in chars:
    print(f'"{l}" appears {text.count(l)} times')
```

```python
# make a vocab
vocab = {word:i for i,word in enumerate(chars)}
vocab
```

```python
# text needs to be a list, not a string
# each element in list is a token
origtext = list(text)
print(text)
print(origtext)
```

```python
# find charecter pairs and merge most frequent
token_pairs = dict()

# loop over tokens
for i in range(len(origtext)-1):

    #create a pair
    pair = origtext[i] + origtext[i+1]

    #increase pair freq
    if pair in token_pairs:
        token_pairs[pair] +=1
    else:
        token_pairs[pair] = 1
token_pairs
```

```python
# find most freq pair
mostFreqPair_idx = np.argmax(list(token_pairs.values()))
mostFreqPair_char = list(token_pairs.keys())[mostFreqPair_idx]
print(f'the most freq char pair is: "{mostFreqPair_char}" with {token_pairs[mostFreqPair_char]} appearences')
```

```python
# update the vocab
vocab[mostFreqPair_char] = max(vocab.values())+1
vocab
```

```python
# replace token pair with one token
newtext = []

#loop through list
i=0
while i<(len(origtext)-1):
    #test wether pair of this and following elements match the newly created pair
    if(origtext[i]+origtext[i+1]) == mostFreqPair_char:
        #append to new version of text
        newtext.append(mostFreqPair_char)
        print(f'added "{mostFreqPair_char}"')

        #skip the next 2 char
        i+=2
    else:
        #this isnt a merged pair, so add this token to list
        newtext.append(origtext[i])
        #move to next char
        i+=1
```

## FUNCTIONISE THE STUFF

```python
## Using functions


def get_pair_stats(text2pair):#text2pair is list of tokens
    token_pairs = dict()
    #loop thorugh tokens
    for i in range(len(text2pair)-1):
        #create a pair
        pair = text2pair[i] + text2pair[i+1]

        #increase pair freq
        if pair in token_pairs:
            token_pairs[pair] +=1
        else:
            token_pairs[pair] = 1
    return token_pairs
```

```python
def update_vocab(token_pairs, vocab):
    #find most freq pair
    idx = np.argmax(list(token_pairs.values()))
    newtok = list(token_pairs.keys())[idx]

    #update voacb
    vocab[newtok] = max(vocab.values())+1
    return vocab,newtok
    
```

```python
def generate_new_token_seq(prevtext,newtoken):
    #init a new text list
    newtext = []

    #loop through list
    i=0
    while i<(len(prevtext)-1):
        #test wether pair of this and following elements match the newly created pair
        if(prevtext[i]+prevtext[i+1]) == newtoken:
            #append to new version of text
            newtext.append(newtoken)
            #skip the next 2 char
            i+=2
        else:
            #this isnt a merged pair, so add this token to list
            newtext.append(prevtext[i])
            #move to next char
            i+=1
    return newtext
```

```python
#re init the vocab

vocab = {word:i for i,word in enumerate(chars)}
print(f'Vocab has {len(vocab)} tokens')
```

```python
## do one iteration

#find and count pairs
pairs = get_pair_stats(origtext)

#update dict
vocab,newtoken = update_vocab(pairs,vocab)

#get a new list of tokens
updated_text = generate_new_token_seq(origtext,newtoken)

print(f'Vocab has {len(vocab)} tokens')
```

```python
# DO A Second ITERATION

pairs = get_pair_stats(updated_text)

#update dict
vocab,newtoken = update_vocab(pairs,vocab)

#get a new list of tokens
updated_text = generate_new_token_seq(updated_text,newtoken)

print(f'Vocab has {len(vocab)} tokens')
```

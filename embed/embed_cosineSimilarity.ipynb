```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics.pairwise import cosine_similarity
import torch
```

```python
## Generate some data
N=30

# data vectors
d1 = np.random.randn(N) + 1 #add an offset for Pearson vs cosine sim
d2 = d1 + np.random.randn(N)  # in the same direction as d1
d3 = -d1 + np.random.randn(N) # opposite dir of d1
```

## Coisne similariy: manual calculation

```python
num = sum(d1*d2)
den = sum(d1*d1) * sum(d2**2)
cossim_manual = num/np.sqrt(den)

print(f'Manula calculation: cosine sim b/w d1 and d2 is {cossim_manual}')
```

```python
# check againts sklearn
# cossim_sklearn = cosine_similarity(d1,d2)

#ghe above wont work as we need to reshape the vectors
cossim_sklearn = cosine_similarity(d1.reshape(1,-1),d2.reshape(1,-1))
print(f'Sklearn cosine sim b/w d1 and d2: {cossim_sklearn}')
```

```python
print(d1.reshape(1,-1).shape)  # vector as a 2d array
print(d1.shape)  # vector

# sklearn and torch needs first format (vectors represneted as 2d array)
```

# comparing with Pearson correlation coeff


```python
#mean center
d1m = d1 - np.mean(d1)
d2m = d2 - np.mean(d2)

#manula calc
num = sum(d1m*d2m)
den = sum (d1m*d1m)*sum(d2m*d2m)
corr_manual = num/np.sqrt(den)

#numpy
corr_np = np.corrcoef(d1,d2)[0,1]

#cosine sim using mean centered data
cossimM_sklearn = cosine_similarity(d1m.reshape(1,-1), d2m.reshape(1,-1))


#print all results
print(f'Manual calculation: Pearson corr b/w d1 and d2 is {corr_manual}')
print(f'Numpy corrcoef(): Pearson corr b/w d1 and d2 is {corr_np}')
print(f'sklearn:  cosine sim  b/w d1m and d2m is {cossimM_sklearn}')
print(f'sklearn:  cosine sim  b/w d1 and d2 is {cossim_sklearn}')

# Pearson correlation coeff is same as cosine similarity if data is mean centered (mean=0)
```
